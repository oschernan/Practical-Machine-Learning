```python
#2016: The instructions and justifications for select2016respondents_df is the same as for the 2015 equivalent DataFrame
#select2016respondents_df has responses that need to be adjusted 
#Q18AGE responses will be changed to mirror 2015 integer response format
select2016respondents_df.replace({'25-34':3,'45-54':5,'55-64':6,'35-44':4,'18-24':2,'65-Over':7,"Don't Know or Refused":8,'Under 23':2,'Under 29':3}, inplace=True)
#Q19GENDER responses will be changed to mirror 2015 integer response format 
select2016respondents_df.replace({'Male':1,'Female':2}, inplace=True)
select2016respondents_df['Q19GENDER'].replace(0, np.nan, inplace=True)
```

```python
#Q21FLY, Q22SJC and Q22OAK have 4 as responses which are not part of dictionary. They will be replaced by NaN
select2016respondents_df['Q21FLY'].replace(4, np.nan, inplace=True)
select2016respondents_df['Q22SJC'].replace(4, np.nan, inplace=True)
select2016respondents_df['Q22OAK'].replace(4, np.nan, inplace=True)
```

```python
#Replace 0 with NaN as in 2015 DataFrame
select2016respondents_df['Q2PURP1'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q2PURP2'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q2PURP3'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q3PARK'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q4BAGS'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q4STORE'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q4FOOD'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q4WIFI'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q5TIMESFLOWN'].replace(0, np.nan, inplace=True) 
select2016respondents_df['Q5FIRSTTIME'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q6LONGUSE'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q16LIVE'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q18AGE'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q20INCOME'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q21FLY'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q22SJC'].replace(0, np.nan, inplace=True)
select2016respondents_df['Q22OAK'].replace(0, np.nan, inplace=True)
```
